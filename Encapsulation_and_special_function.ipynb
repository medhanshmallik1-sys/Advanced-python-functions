```python
class computer:
  def __init__(self):
    self.__maxprice = 900
  def sell(self):
    print("selling Price:{}". format(self.__maxprice))
  def setMAxPrice(self, price):
    self.__maxprice = price
c= computer()
c.sell()
c.setMAxPrice(1000)
c.sell()
```

```python
class point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
  def __str__(self):
    return "({0}, {1})". format(self.x, self.y)
p1 = point(2,3)
print(p1)
```
